### Load Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Load company sales data and store to `sales`

```python
sales = pd.read_csv('company_sales_data.csv')
```

## Use pandas to impute missing valies if there is any

```python
sales.head()
```

```python
sales.info()
```

```python
sales['Shirt'].fillna(sales['Shirt'].mode()[0], inplace=True)
```

## Exercise 1.Read Bags, Top, Shirt and Pant sales data and show it  using a multiline plot


```python
plt.figure(figsize=(12,6))
plt.plot(sales['month_number'], sales['Bags'], marker='o')
plt.plot(sales['month_number'], sales['Top'], marker='o')
plt.plot(sales['month_number'], sales['Shirt'], marker='o')
plt.plot(sales['month_number'], sales['Pant'], marker='o')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Sales data')
plt.legend(labels=['Bag Sales Data', 'Top Sales Data', 'Shirt Sales Data', 'Pant Sales Data'])
plt.show()
```

## Exercise 2: Read Pant and Shirt sales data of each month and show it using a scatter plot


```python
plt.scatter(sales['month_number'], sales['Pant'], color='red')
plt.scatter(sales['month_number'], sales['Shirt'], color='blue')
plt.xlabel('Month number')
plt.ylabel('Number of units sold')
labels = ['Pant sales data','Shirt sales data']
plt.legend(labels=labels)
plt.show()
```

## Exercise 3: Read Jeans and Top product sales data and show it using the bar chart

Note:
-  move x axis of jeans .25 left and Top .25 right and add width
-  plt.bar(sales['month_number']-.25, sales['Jeans'],color='red',width=.30, label="Jeans sales data")

```python

plt.bar(sales['month_number']-.45, sales['Jeans'], color='red', width=.30, label='Jeans sales data')
plt.bar(sales['month_number'], sales['Top'], color='blue', width=.30, label='Top sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Sales data')
plt.legend(labels=['Jeans sales data', 'Top sales data'])
plt.show()
```

## Exercise 4: Read the total profit of each month and show it using the histogram to see the most common profit ranges 
The histogram should look like this:


```python
plt.hist(sales['total_profit'],
        facecolor='blue',
        bins=10,
        edgecolor='blue')
plt.xlabel('Profit in dollar')
plt.title('Profit data')
plt.legend(labels=['Profit data'])
plt.show()
```

## Exercise 5: Calculate `total sale` data for last year for each product and show it using a Pie chart

- salesData=[sales['Bags'].sum(),.......]

```python
sales.info()
```

```python
salesData=[sales['Bags'].sum(), sales['Tie'].sum(),sales['Shirt'].sum(), sales['Jeans'].sum(), sales['Top'].sum(), sales['Pant'].sum()]
```

```python
plt.pie(salesData,labels=['Bags', 'Tie', 'Shirt', 'Jeans', 'Top', 'Pant'],
                                autopct="%.1f%%")
plt.axis('equal')
plt.title('Sales Data')
plt.show()
```

### Exercise 6: Read all product sales data and show it using the stack plot. Save the result as `stack_sales.png`


```python
x = sales['month_number']
y = np.vstack([sales['Bags'], sales['Tie'], sales['Shirt'], sales['Jeans'], sales['Top'], sales['Pant']])
```

```python
labels = ['Bag','Tie', 'Shirt', 'Jeans', 'Top', 'Pant']
colors=['blue', 'orange', 'green', 'red', 'purple', 'brown']
plt.stackplot(x,y, labels=labels, colors=colors)
plt.legend(loc=2)
plt.show()
```
